## Matplotlib의 개요

matplotlib으로 그림을 그릴 때, 아래의 3가지 용어를 알아야 한다.

- Figure : window(창)
- Axes : x-y 좌표계를 가진 Plot을 그릴 수 있는 영역
- Axis : Axes 영역에서의 x축, y축을 가리킴

Figure는 x-y 좌표계를 가진 Plot을 그릴 수 있는 Axes를 1개 이상 가지고 있고, Axes 은 x축, y축을 기본적으로 가지고 있으며, 추가적으로 z축 등을 가질 수 있다.

- matplotlib에서 Figure 와 Axes를 같이 생성할 수 있는 방법중 하나는 `pyplot.subplots` 을 이용하는 것이다.

```python
fig, ax = plt.subplots(nrows=1, ncols=1) # Axes 1개를 가지는 figure를 생성

ax.plot([1, 2, 3, 4], [1, 4, 2, 3]) # Axes에 Plot을 그림

plt.grid(True)
plt.show()
```

- 항상 Figure 와 Axes를 생성해서 사용할 필요는 없다.

- plt.plot 만을 사용해도 충분히 Plot을 그릴 수 있다. (matplotlib.pyplot에서 자동으로 생성하기 때문 또는 이미 생성된 것을 이용하기 때문)

```python
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])
plt.grid(True)
plt.show()
```

- 2개의 방법중 편한 방식으로 사용하면 된다.

- 하지만 2개를 혼용해서 사용하는 것은 추천하지 않는다.

- 전자의 경우 OO-style 이라하고, 후자의 경우 pyplot-style 이라고 한다.

### Plot 그리기

- 데이터 준비하기에서 만들었던 데이터를 사용

```python
fig, ax = plt.subplots() # 1개의 행, 1개의 열을 가진 Figure 객체를 생성

ax.plot(x1, y1) # 선을 그릴 수 있는 기본적인 Plot

plt.show()
```

```python
# 동일한 영역에 겹쳐서 Plot을 그릴 수도 있다.
fig, ax = plt.subplots()

ax.plot(x1, y1)
ax.plot(x2, y2)
ax.scatter(x3, y3)

plt.show()
```

```python
# subplot을 이용해서 하나의 Figure에 개별적으로 그릴 수도 있다.
fig, ax = plt.subplots(3, 1) # (rows, columns) -> 3개의 행, 1개의 열로 이루어진 subplot을 가진 Figure 객체와 Axes 객체를 생성

ax[0].plot(x1, y1)
ax[1].plot(x2, y2)
ax[2].scatter(x3, y3)

plt.tight_layout()
plt.show()
```

```python
# subplot을 이용해서 하나의 Figure에 개별적으로 그릴 수도 있다.
fig, (ax1, ax2, ax3) = plt.subplots(1, 3) # (rows, columns) -> 1개의 행, 3개의 열로 이루어진 subplot을 가진 Figure 객체와 Axes 객체를 생성

ax1.plot(x1, y1)
ax2.plot(x2, y2)
ax3.scatter(x3, y3)

plt.tight_layout()
plt.show()
```

### Plot 커스터마이징

- 아래의 Plot을 이용해서 하나씩 커스터마이징을 할 것이다.

```python
fig, ax = plt.subplots()

ax.plot(x1, y1)
ax.plot(x2, y2)
ax.scatter(x3, y3)

plt.show()
```

- plot에 여러가지 스타일(색깔, 마커, 선 스타일 등)을 지정할 수 있다.

```python
fig, ax = plt.subplots()

# c: color의 약자, ls: linestyle의 약자
ax.plot(x1, y1, c='blue', marker='.', ls='--') # 선의 색깔은 파란색이고, 마커는 .을 사용하고, 선의 스타일은 대쉬이다.
ax.plot(x2, y2, c='red', marker='*', ls=':') # 선의 색깔은 빨간새이고, 마커는 *을 사용하고, 선의 스타일은 점선이다.
ax.scatter(x3, y3)

plt.show()
```

- Plot에 제목을 표시하고 싶을 때, title을 사용한다.

```python
fig, ax = plt.subplots()

ax.plot(x1, y1, c='blue')
ax.plot(x2, y2, c='red')
ax.scatter(x3, y3)

ax.set_title('anatomy of a figure') # (label, ...) -> 'anatomy of a figure'로 title 지정

plt.show()
```

- X축과 Y축에 데이터가 무엇을 나타내는지 표시하고 싶을 때, xlabel 또는 ylabel을 사용한다.

```python
fig, ax = plt.subplots()

ax.plot(x1, y1, c='blue')
ax.plot(x2, y2, c='red')
ax.scatter(x3, y3)

ax.set_title('anatomy of a figure')

ax.set_xlabel('X data') # (xlabel, ...) -> "X data"로 label 지정
ax.set_ylabel('Y data') # (ylabelm, ...) -> "Y data"로 label 지정

plt.show()
```

- X축 데이터 또는 Y축 데이터의 범위를 지정하고 싶을 때, xlim 또는 ylim을 사용한다.

```python
fig, ax = plt.subplots()

ax.plot(x1, y1, c='blue')
ax.plot(x2, y2, c='red')
ax.scatter(x3, y3)

ax.set_xlabel('X data')
ax.set_ylabel('Y data')
ax.set_title('anatomy of a figure')

ax.set_xlim(0, 4) # (left, right) -> X축의 경우 0 ~ 4 사이로 범위 지정
ax.set_ylim(-1, 8) # (left, right) -> Y축의 경우 -1 ~ 8 사이로 범위 지정

plt.show()
```

- X축 또는 Y축 데이터의 간격을 조절하고 싶을 때, xticks 또는 ytickes를 사용한다.

- X축과 Y축 데이터의 간격에 레이블을 표시하고 싶을 때, xticklabels 또는 yticklabels 사용한다. (시각화 예시 부분에서 설명)

```python
fig, ax = plt.subplots()

ax.plot(x1, y1, c='blue')
ax.plot(x2, y2, c='red')
ax.scatter(x3, y3)

ax.set_xlim(0, 4)
ax.set_ylim(-1, 8)
ax.set_xlabel('X data')
ax.set_ylabel('Y data')
ax.set_title('anatomy of a figure')

ax.set_xticks(np.arange(0, 4.25, step=0.25), minor=True) # X축의 간격을 0.25로 지정
ax.set_yticks(np.arange(-1, 8.25, step=0.25), minor=True) # Y축의 간격을 0.25로 지정
ax.minorticks_on()

plt.show()
```

- 범례를 표시하고 싶을 때, legend 를 사용한다.

```python
fig, ax = plt.subplots()

ax.plot(x1, y1, c='blue', label='Blue signal') # Plot이 사용할 범례를 label에 지정
ax.plot(x2, y2, c='red', label='Red signal') # Plot이 사용할 범례를 label에 지정
ax.scatter(x3, y3)

ax.set_xticks(np.arange(0, 4.25, step=0.25), minor=True)
ax.set_yticks(np.arange(-1, 8.25, step=0.25), minor=True)
ax.minorticks_on()
ax.set_xlim(0, 4)
ax.set_ylim(-1, 8)
ax.set_xlabel('X data')
ax.set_ylabel('Y data')
ax.set_title('anatomy of a figure')

ax.legend() # 범례 표시

plt.show()
```

### plt.plot

- 가장 기본적인 선을 그리를 Line Plot

```python
# Prepare data
x = np.linspace(0, 10, 100)
y = np.cos(x)

# Create plot
fig, ax = plt.subplots()

# plot
ax.plot(x, y)

# Show plot
plt.show()
```

### plt.scatter plot

- 두 개 변수 간의 관계를 보고 싶을 때 사용하는 Scatter plot

```python
from sklearn.datasets import load_iris # 붓꽃데이터 로드

# Prepare Data
iris = load_iris()
x = iris.data[:, 0] # Sepal Length
y = iris.data[:, 1] # Sepal Width
target = iris.target # Target

# Create plot
fig, ax = plt.subplots()

# plot
ax.scatter(x, y, c=target) # Target 별로, Sepal Length, Sepal Width간의 관계 확인

# Show plot
plt.show()
```

### plt.bar

- 카테고리 별 빈도 또는 합계 등을 비교하는데 유용한 그래프

```python
import seaborn as sns

# prepare data
titanic = sns.load_dataset('titanic')
labels = ['feamle', 'male']
survived_0 = titanic.groupby(['sex', 'survived']).size().unstack()[0] 
survived_1 = titanic.groupby(['sex', 'survived']).size().unstack()[1]

x = np.arange(len(labels))
width = 0.35

# Create plot
fig, ax = plt.subplots()

# plot
bc1 = ax.bar(x - width/2, survived_0, width, label='0')
bc2 = ax.bar(x + width/2, survived_1, width, label='1')

# Customize plot
ax.set_ylabel('Count')
ax.set_title('survived according sex')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(title='survived')

ax.bar_label(bc1, padding=3)
ax.bar_label(bc2, padding=3)

# Show plot
fig.tight_layout()
plt.show()
```

### plt.pie

- 카테고리 별 상대적인 비교를 하는데 유용한 그래프

```python
import seaborn as sns

# Prepare data
titanic = sns.load_dataset('titanic')
labels = ['female', 'male']
sizes = titanic.groupby(by='sex').size().to_numpy()
explode = (0, 0.1)

# Create plot
fig, ax = plt.subplots()

# plot
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.2f%%',
       shadow=True, startangle=90)

# Customize plot
ax.axis('equal')
ax.set_title('Sex ratio')

# Show plot
fig.tight_layout()
plt.show()
```

### plt.hist

- 분포도를 확인할 때 유용한 그래프

```python
import seaborn as sns

# Prepare data
titanic = sns.load_dataset('titanic')
x = titanic.age

# Create plot
fig, ax = plt.subplots()

# plot
ax.hist(x, bins=25)

# Customize plot
ax.set_title('Frequency by age')
ax.set_ylabel('Frequency')
ax.set_xlabel('Age')
ax.grid(True)

# Show plot
plt.show()
```

### plt.boxplot

- 1사분위수, 2사분위수, 3사분위수, 4사분위수, 아웃라이어를 확인할 때 유용한 그래프

```python
import seaborn as sns

# Prepare data
titanic = sns.load_dataset('titanic')
x = titanic.age[titanic.age.isnull() == False].values

# Create plot
fig, ax = plt.subplots()

# plot
ax.boxplot(x)

# Customize plot
ax.set_xticklabels(['age'])
ax.set_title('detect outlier about age')

# Show plot
plt.show()
```

### Axes3d.plot_surface

- 3차원 데이터를 그릴 때 유용한 그래프 (2D Axes가 아닌 3D Axes를 이용해서 그림)

```python
# Prepare data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y =  np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Create Plot
fig, ax = plt.subplots(subplot_kw={'projection': '3d'}) # Axed3D

# plot
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', lw=0, antialiased=False)

# Customize plot
ax.set_title("3D Surface Plot")
ax.set_zlim(-1.01, 1.01)
ax.set_zticks(np.linspace(-1.01, 1.01, 10))
ax.zaxis.set_major_formatter('{x:.02f}')
fig.colorbar(surf, shrink=0.5, aspect=5, pad=0.1)

# Show plot
plt.show()
```
